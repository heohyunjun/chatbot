```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer('jhgan/ko-sroberta-multitask')

sentences = ['안녕하세요' , '한국어 문장 임베딩을 위한 버트 모델입니다']

embeddings = model.encode(sentences)

print(embeddings)
```

```python
# 챗봇 컬럼에 null있는가 확인
# False : null 없음
# True : null 있음
df['챗봇'].isna()
```

```python
# null 이 아닌것 , not = ~
~df['챗봇'].isna()
```

```python
df = df[~df['챗봇'].isna()]
df.head()
```

```python
# 인코딩 해보기
ex_ = df.loc[0, '유저']
print(ex_)
```

```python
# 인코딩
model.encode(ex_)
```

```python
# 전체 데이터셋 인코딩

df['embedding'] = pd.Series([[]] * len(df))
```

```python
df
```

```python
pd.Series([[]])
```

```python
# 전체 데이터셋 인코딩
df['embedding'] = df['유저'].map(lambda x : list(model.encode(x)))
```

```python
# 저장해두기
df.to_csv('/content/drive/MyDrive/Colab Notebooks/챗봇/chat_text.csv',index =False)
```

```python
user_text = '머리가 아프고 힘들다'
```

```python
embedding = model.encode(user_text)
```

```python
embedding
```

원리
- embedding 컬럼값들과 답변을 원하는 uesr_text 인코딩값사이 문서 유사도를 구해서 가장 높은 유사도를 가진 놈을 선택

```python
df['distance'] = df['embedding'].map(lambda x : cosine_similarity([embedding], [x]).squeeze())
```

```python
df['distance']
```

```python
answer = df.loc[df['distance'].idxmax()]
```

```python
answer
```

```python
print('구분', answer['구분'])
print('유사 질문', answer['유저'])
print('챗봇 답변', answer['챗봇'])
print('유사도', answer['distance'])
```
